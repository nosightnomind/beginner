# 인공지능의 분야와 기술

 ## 인공지능 

  인공지능의 개념은 지적작업을 자동화하려는 노력에서 시작됬습니다. 초기의 인공지능은 하드 코닝 방식을 이용한 기호주의 인공지능으로 논리적인 영역에서는 강점을 보였지만 번역과 음성 이미지 등 모호한 영역에서 부족과 하드웨어의 한계에 부딪힙니다. 하지만 하드웨어의 발전과 인간의 뇌를 모방한 인지컴퓨팅 기술은 입력과 출력값의 사이의 규칙을 추출하는 머신러닝과 그중 올바른 값을 찾는 딥 러닝의 발전으로 인공지능은 국방 금융 의학 제조산업 등 다양한 분야에서 활용되고 있습니다.

## 생활 속 인공지능

+ 파파고 

  파파고는 언어영역에서 인공지능의 예시로 음성, 사진 등을 통해 문자를 인식하여 자신이 원하는 언어로 번역을 해주는 역할을 합니다. 기존의 언어 번역이 문장을 구 또는 단어로 나눈 뒤 통계적 모델에 기반하여 번역하던 것과 달리 인공신경망 번역을 통해 문장의 전체 맥락을 파악한 뒤 어순 의미 문맥별 의미 차이를 스스로 수정하여 번역결과를 내놓습니다.   
  
     
+  인공지능 피부관리기 

  이미지 분야에서 인공지능의 예시로 한 번의 촬영으로 모공 주름 등 6가지 피부 항목을 분석하고 사용자에게 맞는 제품 및 케어를 추천하는 제품입니다. 이미지인식은 노이즈를 픽셀로 데이터를 인식하기 떄문에 노이즈를 판별하는 것이 중요한 역활로서 사물의 특징을 잡고 추상화하는 기술이 필요로 합니다.    

   
+ AI 스피커 

  음성영역에서 가장 대표되며 주변에서 심심찮게 보이는 기술로 AI 스피커가 있습니다. 음성인식은 소리를 문자로 변환에서 문맥에 맞는 행동을 한다는 점에서 언어분석 영역과 밀접한 관계를 갖고 있습니다. 주위소리에서 올바른 소리를 인식하고 사투리 말투 축약어 등 언어의 한계 속에서 많은 데이터를 요구합니다. IoT, 자동차 가전제품 등 다양한 분야에서 활용될 여지가 많습니다.   

   
+ 자율주행 자동차

  자율주행 자동차는 인지, 판단, 제어 3단계에 맞춰 이루어집니다. 인지단계에서 카메라 레이더 GPS등 많은 정보를 인식하고 수집해야 합니다. 이 단계에서 순간적으로 많은 정보를 빠르고 정확하게 보낼 수 있는 기술력이 필요로 합니다. 판단단계에서 인지정보를 바탕으로 한 주행 전략을 결정하고 제어단계에서 실제로 차량의 주행을 시작합니다. 딥러닝의 적용으로 시스템에 필요한 동적, 정적환경 요소 검출, 다른 운전자의 움직임 예측, 승차감의 개선 등 자율주행자동차의 발전 가능성은 많습니다.
